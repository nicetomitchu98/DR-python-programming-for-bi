## Task 1 - list of dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

### a) Filtering invalid transactions

```python
#simple filter to remove transactions with negative price or quantity
#pointer to the another list object in memory, so it does not modify the original list of transactions
filtered_transactions = [t for t in transactions if t["price"]>0 and t["quantity"]>=0]

print(f"There are {len(filtered_transactions)} valid transactions and the breakdown is below:")
for transaction in filtered_transactions:
    print(f"Transaction ID {transaction['id']}: Price = {transaction['price']}, Quantity = {transaction['quantity']}")
```

### b) Filtering invalid transactions

```python
#Calculating the total and adding it as a new key-value pair in the transaction dictionary
for transaction in filtered_transactions:
    transaction["total"] = transaction["price"] * transaction["quantity"]

#Print nice and tidy 
for transaction in filtered_transactions:
    print(f"Transaction ID {transaction['id']}: Total = {transaction['total']}")
```

### c) Grouping by region

```python
#initializing an empty dictionary to store the grouped regions
grouped_regions = {}

#grouping the transactions by region
for transaction in filtered_transactions:

    #placeholder to store the region of the current transaction
    region = transaction["region"]

    #if region was not appended earlier, create a new entry with current trasaction
    if region not in grouped_regions:
        grouped_regions[region] = [transaction]
    
    #otherwise, just append the transaction to the exisiting key in the dictionary
    else:
        grouped_regions[region].append(transaction)

#quick sanity check
print(grouped_regions)


#create a new dictionary to store the total revenue for each region
#later i will utilize grouped_regions so dont want to modify it 

total_revenue_regions= {}
#calculating the total revenue for each region and updating the dictionary 
# with the total revenue instead of the list of transactions
for region, transactions in grouped_regions.items():
    total_revenue = sum(t["total"] for t in transactions)
    total_revenue_regions[region] = total_revenue

#final
print(total_revenue_regions)
```

### d) Region with highest total revenue

```python
#Initialize container for the maximum total of the region
region_max=0

#going though all transactions grouped by region
for region in grouped_regions:

    #summing over all transactions in the region
    region_total = sum(t["total"] for t in grouped_regions[region])

    #conditional statmeent if the total of the current region is higher
    #than previously set maximum
    if region_total>region_max:
        region_max = region_total
        max_region = region

print(f"The region with the highest total sales is {max_region} with a total of {region_max}.")
```

## Task 2 - creating a function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

```python
def summarize_sales(sales):

    output_dict = {}

    quantity = len(sales)
    total = sum(sales)

    output_dict["total"] = total

    #well, we cannot calculate the average if either nominator or denominator is zero
    if quantity != 0 and total != 0:
        average = total / quantity
    else:
        raise ValueError("Quantity and total must be non-zero to calculate average.")
    
    output_dict["average"] = average

    #iterating through the sales to find the best day
    best_sale=max(sales)
    best_day=sales.index(best_sale) # we find the index of the best sale, which corresponds to the day 
    #number (starting from 0)
    

    # we add 1 to the index becuase otherwise the day number would start from 0
    #it is more intiutive to say that the best day is e.g. day 1, not day 0
    best_day += 1 


    output_dict["best_day"] = best_day

    #same as best day, but for the worst day
    worst_sale=min(sales)
    worst_day=sales.index(worst_sale)
    worst_day += 1
    output_dict["worst_day"] = worst_day

    above_average_count = [i for i, sale in enumerate(sales) if sale > average]

    output_dict["above_average_count"] = len(above_average_count)

    return output_dict

summarize_sales(sales)
```

### c) Growth adjustment

The logic is to first isolate the revenue of 2024 by taking the last row with index -1 and all the columns using : operator

We will get an error if we try to print the shape of growth_rate, because it is a list, not a numpy array, so it does not have the attribute shape

Once the shapes are compatible, we can apply simple multiplication

**The multiplication works well even without converting growth_rate to a NumPy array**, because NumPy supports element-wise operations with array-like objects such as Python lists, as long as their shapes are compatible. In this case, both have length 4, so the operation proceeds without broadcasting issues. 

Converting growth_rate to a NumPy array is not required here, but is safer, it allows access to NumPy methods and allows memory efficiency if computations are longer

```python
growth_rate = [1.05, 1.03, 1.08, 1.02]

revenue_2024 = revenue[-1,:] 


print(f"Shape of revenue_2024: {revenue_2024.shape}")

#will result in error 
#print(f"Shape of growth_rate: {growth_rate.shape}") 

#we need to convert the growth_rate to a numpy array to be able to perform element-wise multiplication
growth_rate = np.array(growth_rate)
print(f"Shape of growth_rate: {growth_rate.shape}") 


revenue_2024= revenue_2024 * growth_rate
print(f"Adjusted revenue for 2024: {revenue_2024}")
```

### d) Good quarters

I apply the step by step code production for the given task. basically one could stop already at the first 
sub-point where we get the revenues above the threshold, but in my opinion it is not very descriptive.

For data analysis it is preferred to identify the numbers with the mapped time entries, i.e years and quarters

```python
revenue_threshold = 250

#when we are asked to use the boolean indexing, lets first create a boolean array that indicates 
#which elements of the revenue array are above the threshold
boolean_mask = revenue > revenue_threshold
print(boolean_mask)

#we can then use this boolean mask to filter the 
#revenue array and get only the values that are above the threshold
above_threshold_revenue = revenue[boolean_mask]
print(above_threshold_revenue)
```

```python
#we can also identify the year and quarter for each of the above threshold revenues using the same boolean mask
above_threshold_years = np.where(boolean_mask)[0]  # get the row indices (years)
above_threshold_quarters = np.where(boolean_mask)[1]  # get the column indices
print(f"Years for above threshold revenue: {above_threshold_years}")
print(f"Quarters for above threshold revenue: {above_threshold_quarters}")
```

```python
#we can represent teh combined info in format of dictionary, where the keys 
#are the above threshold revenues and the values are tuples of (year, quarter)
above_threshold_info = {}
for i in range(len(above_threshold_revenue)):
    revenue_value = above_threshold_revenue[i]
    year = above_threshold_years[i]
    quarter = above_threshold_quarters[i]
    above_threshold_info[revenue_value] = (year, quarter)
print(f"Above threshold revenue info: {above_threshold_info}")
```

```python
#and if we go even further, then we can mark years, i.e. year 0=2022, year 1=2023, year 2=2024, 
#and quarters as well, i.e. quarter 0=Q1, quarter 1=Q2, quarter 2=Q3, quarter 3=Q4
above_threshold_info_detailed = {}
for i in range(len(above_threshold_revenue)):
    revenue_value = above_threshold_revenue[i]
    year = above_threshold_years[i]
    quarter = above_threshold_quarters[i]
    year_label = f"202{year+2}"  # since year 0 corresponds to 2022
    quarter_label = f"Q{quarter+1}"  # since quarter 0 corresponds to Q1
    above_threshold_info_detailed[revenue_value] = (year_label, quarter_label)

print(f"Above threshold revenue info pretty: {above_threshold_info_detailed}")
```
